```python
from sqlalchemy import *
from sqlalchemy.engine import create_engine
from sqlalchemy.schema import *

engine = create_engine('druid://broker:8082/druid/v2/sql?header=true')

wikipedia = Table('wikipedia', MetaData(bind=engine), autoload=True)
print(select([func.count('*')], from_obj=wikipedia).scalar())
```

```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, DateTime

Base = declarative_base()

class Wikipedia(Base):
    __tablename__ = 'wikipedia'
    
    time = Column('__time', DateTime)
    channel = Column(String)
    cityName = Column(String)
    page = Column(String, primary_key=True)
```

```python
from sqlalchemy.orm import sessionmaker

Session = sessionmaker(bind=engine)
session = Session()
for instance in session.query(Wikipedia):
    print(instance.channel, instance.page, instance.time)
```
